```python
import pandas as pd

df = pd.read_csv(RAW_PATH / "01_fund_master.csv")

print("Original Shape:", df.shape)

# Remove duplicates
df = df.drop_duplicates()

# AMFI code as string
df["amfi_code"] = df["amfi_code"].astype(str)

# Convert date
df["launch_date"] = pd.to_datetime(
    df["launch_date"],
    errors="coerce"
)

# Clean text columns
text_cols = [
    "fund_house",
    "scheme_name",
    "category",
    "sub_category",
    "plan",
    "benchmark",
    "fund_manager",
    "risk_category",
    "sebi_category_code"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

# Save cleaned file
df.to_csv(
    PROCESSED_PATH / "01_fund_master_cleaned.csv",
    index=False
)

print("Saved Successfully")
print(df.shape)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(RAW_PATH / "03_aum_by_fund_house.csv")

print("Original Shape:", df.shape)

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Convert date column
df["date"] = pd.to_datetime(
    df["date"],
    errors="coerce"
)

# 3. Clean fund house names
df["fund_house"] = (
    df["fund_house"]
    .astype(str)
    .str.strip()
)

# 4. Validate AUM values
invalid_aum = df[
    (df["aum_lakh_crore"] <= 0) |
    (df["aum_crore"] <= 0)
]

print("Invalid AUM Records:", len(invalid_aum))

# 5. Validate scheme counts
invalid_schemes = df[
    df["num_schemes"] <= 0
]

print("Invalid Scheme Counts:", len(invalid_schemes))

# 6. Sort data
df = df.sort_values(
    by=["fund_house", "date"]
)

# 7. Save cleaned file
df.to_csv(
    PROCESSED_PATH / "03_aum_by_fund_house_cleaned.csv",
    index=False
)

print("Cleaned Shape:", df.shape)
print("Saved Successfully")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(RAW_PATH / "04_monthly_sip_inflows.csv")

print("Original Shape:", df.shape)

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Convert month column
df["month"] = pd.to_datetime(
    df["month"],
    format="%Y-%m",
    errors="coerce"
)

# 3. Validate numeric columns
numeric_cols = [
    "sip_inflow_crore",
    "active_sip_accounts_crore",
    "new_sip_accounts_lakh",
    "sip_aum_lakh_crore"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )

# 4. Fill YoY Growth Nulls
df["yoy_growth_pct"] = (
    df["yoy_growth_pct"]
    .fillna(0)
)

# 5. Validate values
invalid_rows = df[
    (df["sip_inflow_crore"] <= 0) |
    (df["active_sip_accounts_crore"] <= 0) |
    (df["new_sip_accounts_lakh"] <= 0)
]

print("Invalid Records:", len(invalid_rows))

# 6. Sort by month
df = df.sort_values("month")

# 7. Save cleaned file
df.to_csv(
    PROCESSED_PATH / "04_monthly_sip_inflows_cleaned.csv",
    index=False
)

print("Cleaned Shape:", df.shape)
print("Saved Successfully")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(RAW_PATH / "05_category_inflows.csv")

print("Original Shape:", df.shape)

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Convert month column
df["month"] = pd.to_datetime(
    df["month"],
    format="%Y-%m",
    errors="coerce"
)

# 3. Clean category names
df["category"] = (
    df["category"]
    .astype(str)
    .str.strip()
)

# 4. Convert inflow to numeric
df["net_inflow_crore"] = pd.to_numeric(
    df["net_inflow_crore"],
    errors="coerce"
)

# 5. Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# 6. Sort data
df = df.sort_values(
    by=["month", "category"]
)

# 7. Save cleaned file
df.to_csv(
    PROCESSED_PATH / "05_category_inflows_cleaned.csv",
    index=False
)

print("\nCleaned Shape:", df.shape)
print("Saved Successfully")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(RAW_PATH / "06_industry_folio_count.csv")

print("Original Shape:", df.shape)

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Convert month column
df["month"] = pd.to_datetime(
    df["month"],
    format="%Y-%m",
    errors="coerce"
)

# 3. Convert folio columns
folio_cols = [
    "total_folios_crore",
    "equity_folios_crore",
    "debt_folios_crore",
    "hybrid_folios_crore",
    "others_folios_crore"
]

for col in folio_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )

# 4. Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# 5. Validate folio values
invalid_rows = df[
    (df["total_folios_crore"] <= 0)
]

print("\nInvalid Records:", len(invalid_rows))

# 6. Sort data
df = df.sort_values("month")

# 7. Save cleaned file
df.to_csv(
    PROCESSED_PATH / "06_industry_folio_count_cleaned.csv",
    index=False
)

print("\nCleaned Shape:", df.shape)
print("Saved Successfully")
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(RAW_PATH / "09_portfolio_holdings.csv")

print("Original Shape:", df.shape)

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Convert AMFI code
df["amfi_code"] = df["amfi_code"].astype(str)

# 3. Convert portfolio date
df["portfolio_date"] = pd.to_datetime(
    df["portfolio_date"],
    errors="coerce"
)

# 4. Clean text columns
text_cols = [
    "stock_symbol",
    "stock_name",
    "sector"
]

for col in text_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.upper()
    )

# 5. Convert numeric columns
numeric_cols = [
    "weight_pct",
    "market_value_cr",
    "current_price_inr"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )

# 6. Validate values
invalid_rows = df[
    (df["weight_pct"] <= 0) |
    (df["market_value_cr"] <= 0) |
    (df["current_price_inr"] <= 0)
]

print("Invalid Records:", len(invalid_rows))

# 7. Sort data
df = df.sort_values(
    by=["amfi_code", "weight_pct"],
    ascending=[True, False]
)

# 8. Save cleaned file
df.to_csv(
    PROCESSED_PATH / "09_portfolio_holdings_cleaned.csv",
    index=False
)

print("Cleaned Shape:", df.shape)
print("Saved Successfully")
```
